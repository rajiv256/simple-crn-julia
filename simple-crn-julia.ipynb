```julia
using DifferentialEquations
```

```julia
function hjelmfelt!(du, u, p, t)
    k1 = 100.0
    km1 = 1.0
    k2 = 5e4
    km2 = 1.0
    k3 = 5e4
    km3 = 1.0
    k4 = 1.0
    km4 = 100

    
    J1 = k1*u[5] - km1*u[5]*u[1]
    J2 = k2*u[1]*u[4] - km2*u[3]
    J3 = k3*u[2]*u[3] - km3*u[4]
    J4 = k4*u[2] - km4

    # u = [X1, X3, A, B, C]
    du[1] = J1 - J2
    du[2] = -J3 - J4
    du[3] = J2 - J3
    du[4] = -du[3]
end 
```

```julia
u0 = [0.0;0.0;0.1;0.9;0.9]
tspan = (0.0, 0.1)
prob = ODEProblem(hjelmfelt!, u0, tspan)
sol = solve(prob)
using Plots

plot(sol, vars=[3], label='A')
# plot(sol, vars=[4], label='B')
# plot(sol, vars=[5], label='C')
```
